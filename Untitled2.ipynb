```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('mtcars (1) (1).csv')
df
```

```python
#histogram
range=(10,40)
bins=10
plt.hist(df['mpg'],bins,range,color='orange',rwidth=0.8)
plt.xlabel('mpg')
plt.ylabel('no. of frequency')
plt.title('my histogram')
plt.show()
```

```python
#scatter plot
plt.scatter(df['wt'],df['mpg'],color='green',marker='*')
plt.title('scatter plot')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
from pandas._libs.hashtable import value_count
#box plot
temp=df['wt'].value_counts()
print(temp)
temp.plot.bar()
plt.show()
```
